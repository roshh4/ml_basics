```python
x = data.drop(['MEDV'], axis =1)
```

```python
y = data['MEDV']
```

```python
kf = KFold(random_state=40,shuffle=True, n_splits=3)
```

```python
for train_index, test_index in kf.split(x):
    print(test_index)
```

```python
for train_index, test_index in kf.split(x):
    x_train, x_test, y_train, y_test = (x.iloc[train_index, :],x.iloc[test_index, :], y[train_index], y[test_index])
```

```python
score = []
model = LinearRegression()
model
scaler = StandardScaler()
```

```python
for train_index, test_index in kf.split(x):
    x_train, x_test, y_train, y_test = (x.iloc[train_index, :],x.iloc[test_index, :], y[train_index], y[test_index])
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.fit_transform(x_test)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score.append(r2_score(y_test.values, y_pred))
score
```

```python
lr = LinearRegression()
sc = StandardScaler()
estimator = Pipeline([("scaler", sc), ("regression", lr)])
estimator
```

```python
predictions = cross_val_predict(estimator, x, y, cv=kf)
predictions
```

```python
r2_score(y,predictions)
```

```python
np.mean(score)
```

```python
alphas = np.geomspace(1e-9,1e0,num=10)
alphas
```

```python
scores = []
coeffs = []
for alpha in alphas:
    las = Lasso(alpha= alpha, max_iter=100000)
    estimator = Pipeline([("scaler", sc), ("lasso_regression", las)])
    prediction = cross_val_predict(estimator, x, y, cv = kf)
    scores.append(r2_score(y,prediction))
scores
```

```python
list(zip(alphas,scores))
```

```python
pf = PolynomialFeatures(degree=2)
scores = []

alphas = np.geomspace(0.001, 10, 5)
for alpha in alphas:
    las = Lasso(alpha=alpha, max_iter=100000)
    estimator = Pipeline([
        ("make_higher_degree", pf),
        ("scaler", sc),
        ("lasso_regression", las)])
    predictions = cross_val_predict(estimator, x, y, cv = kf)
    score = r2_score(y, predictions)
    
    scores.append(score)
scores
```

```python
pf = PolynomialFeatures(degree=2)
scores = []

alphas = np.geomspace(4, 20, 20)
for alpha in alphas:
    ridge = Ridge(alpha=alpha, max_iter=100000)

    estimator = Pipeline([
        ("polynomial_features", pf),
        ("scaler", sc),
        ("ridge_regression", ridge)])

    predictions = cross_val_predict(estimator, x, y, cv = kf)
    score = r2_score(y, predictions)
    scores.append(score)
scores
```

```python
from sklearn.model_selection import GridSearchCV
estimator = Pipeline([("polynomial_features", PolynomialFeatures()), ("scaler", StandardScaler()), ("ridge_regression", Ridge())])
params = {'polynomial_features__degree': [1,2,3,], 'ridge_regression__alpha':np.geomspace(4,20,20)}
grid = GridSearchCV(estimator, params, cv=kf)
```

```python
grid.fit(x,y)
grid.best_score_, grid.best_index_
```

```python
y_predict = grid.predict(x)
r2_score(y,y_predict)
```
