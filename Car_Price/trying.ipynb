```python
columns=['price', 'fueltype', 'aspiration','carbody', 'drivewheel','wheelbase', 'brand',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'carlength','carwidth','citympg','highwaympg']
```

```python
data.columns.tolist()
```

```python
selected = data[columns]
selected.info()
```

```python
categorical_columns=[key for key, value in selected.dtypes.items()  if value=='O']
categorical_columns
```

```python
numeric_columns=list(set(columns)-set(categorical_columns))
numeric_columns
```

```python
x = selected.drop('price',axis=1)
```

```python
x
```

```python
y = selected.price
```

```python
y
```

```python
for column in  categorical_columns:
    print("column name:", column)
    print("value_count:")
    print( x[column].value_counts())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False)
one_hot = ColumnTransformer(
    transformers=[
        ("one_hot", one_hot_encoder, categorical_columns)
    ],
    remainder="passthrough"
)
x = one_hot.fit_transform(x)
print(type(x))
```

```python
names=one_hot.get_feature_names_out()
names
```

```python
colunm_names=[name[name.find("_")+1:] for name in  [name[name.find("__")+2:] for name in names]]
colunm_names
```

```python
print(x)
```

```python
df=pd.DataFrame(data=x,columns=colunm_names)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss
x_train
```

```python
x_train=ss.fit_transform(x_train)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
x_test = ss.transform(x_test)
```

```python
x_test
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_predicted)
mse
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test, y_predicted)
```
